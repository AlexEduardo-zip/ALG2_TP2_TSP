```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('resultados_com_fator.csv')

df_filtered = df[df['algoritmo_utilizado'] != 'Branch and Bound']

df_filtered.head()

df = df_filtered

df.head()
```

```python
tat_df = df[df['algoritmo_utilizado'] == 'Twice Around the Tree']
chr_df = df[df['algoritmo_utilizado'] == 'Christofides']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['tempo_de_Execucao'] = pd.to_numeric(df['tempo_de_Execucao'], errors='coerce')

df_christofides = df[df['algoritmo_utilizado'] == 'Christofides']
df_twice_around = df[df['algoritmo_utilizado'] == 'Twice Around the Tree']

plt.figure(figsize=(12, 6))

sns.lineplot(x='nome_do_problema', y='tempo_de_Execucao', data=df_christofides, label='Christofides', marker='o', color='r')

sns.lineplot(x='nome_do_problema', y='tempo_de_Execucao', data=df_twice_around, label='Twice Around the Tree', marker='o', color='b')

plt.title('Comparao de Tempo de Execuo entre Christofides e Twice Around the Tree')
plt.xticks(rotation=90)  
plt.ylim(0, 60)  

plt.xlabel('Problema')
plt.ylabel('Tempo de Execuo (segundos)')

plt.tight_layout()  
plt.legend()  
plt.show()
```

```python
df['quantidade_memoria_em_mb'] = pd.to_numeric(df['quantidade_memoria_em_mb'], errors='coerce')

df_christofides = df[df['algoritmo_utilizado'] == 'Christofides']
df_twice_around = df[df['algoritmo_utilizado'] == 'Twice Around the Tree']

plt.figure(figsize=(12, 6))

sns.lineplot(x='nome_do_problema', y='quantidade_memoria_em_mb', data=df_christofides, label='Christofides', marker='o', color='r')

sns.lineplot(x='nome_do_problema', y='quantidade_memoria_em_mb', data=df_twice_around, label='Twice Around the Tree', marker='o', color='b')

plt.title('Comparao de Memria Usada entre Christofides e Twice Around the Tree')
plt.xticks(rotation=90)  
plt.ylim(0, 150)  

plt.xlabel('Problema')
plt.ylabel('Memria Usada (MB)')

plt.tight_layout()  
plt.legend()  
plt.show()
```

```python
df['fator_aproximacao'] = pd.to_numeric(df['fator_aproximacao'], errors='coerce')

df_christofides = df[df['algoritmo_utilizado'] == 'Christofides']
df_twice_around = df[df['algoritmo_utilizado'] == 'Twice Around the Tree']

plt.figure(figsize=(12, 6))

sns.lineplot(x='nome_do_problema', y='fator_aproximacao', data=df_christofides, label='Christofides', marker='o', color='r')

sns.lineplot(x='nome_do_problema', y='fator_aproximacao', data=df_twice_around, label='Twice Around the Tree', marker='o', color='b')

plt.title('Comparao do Fator de Aproximao entre Christofides e Twice Around the Tree')
plt.xticks(rotation=90)   

plt.xlabel('Problema')
plt.ylabel('Fator de Aproximao')

plt.tight_layout()   
plt.legend()   
plt.show()
```

```python
df['tempo_de_Execucao'] = pd.to_numeric(df['tempo_de_Execucao'], errors='coerce')
df['quantidade_Nos'] = pd.to_numeric(df['quantidade_Nos'], errors='coerce')

df_algoritmo1 = df[df['algoritmo_utilizado'] == 'Christofides']
df_algoritmo2 = df[df['algoritmo_utilizado'] == 'Twice Around the Tree']

plt.figure(figsize=(12, 6))

sns.lineplot(x='quantidade_Nos', y='tempo_de_Execucao', data=df_algoritmo1, label='Christofides', marker='o', color='r')

sns.lineplot(x='quantidade_Nos', y='tempo_de_Execucao', data=df_algoritmo2, label='Twice Around the Tree', marker='o', color='b')

plt.title('Comparao de Tempo de Execuo entre Christofides e Twice Around the Tree por Quantidade de Ns')
plt.ylim(0, 1500)  

plt.xlabel('Quantidade de Ns')
plt.ylabel('Tempo de Execuo (segundos)')

plt.tight_layout()
plt.legend()
plt.show()
```
