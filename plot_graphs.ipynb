```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o arquivo CSV
df = pd.read_csv('resultados_com_fator.csv')

# Filtrar o DataFrame para remover as linhas onde o algoritmo seja 'Branch and Bound'
df_filtered = df[df['algoritmo_utilizado'] != 'Branch and Bound']

# Verificar as primeiras linhas do DataFrame filtrado
df_filtered.head()

df = df_filtered

# Verificar as primeiras linhas do dataframe
df.head()
```

```python
# Filtrar os dados para os dois algoritmos
tat_df = df[df['algoritmo_utilizado'] == 'Twice Around the Tree']
chr_df = df[df['algoritmo_utilizado'] == 'Christofides']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Garantir que a coluna 'tempo_de_Execucao' seja numrica
df['tempo_de_Execucao'] = pd.to_numeric(df['tempo_de_Execucao'], errors='coerce')

# Filtrar os dados para os algoritmos 'Christofides' e 'Twice Around the Tree'
df_christofides = df[df['algoritmo_utilizado'] == 'Christofides']
df_twice_around = df[df['algoritmo_utilizado'] == 'Twice Around the Tree']

# Plotando os grficos para ambos os algoritmos no mesmo grfico
plt.figure(figsize=(12, 6))

# Plot para o algoritmo 'Christofides'
sns.lineplot(x='nome_do_problema', y='tempo_de_Execucao', data=df_christofides, label='Christofides', marker='o', color='r')

# Plot para o algoritmo 'Twice Around the Tree'
sns.lineplot(x='nome_do_problema', y='tempo_de_Execucao', data=df_twice_around, label='Twice Around the Tree', marker='o', color='b')

# Ajustando o ttulo e os limites do grfico
plt.title('Comparao de Tempo de Execuo entre Christofides e Twice Around the Tree')
plt.xticks(rotation=90)  # Rotaciona os rtulos no eixo X para melhor visualizao
plt.ylim(0, 60)  # Limitar o eixo Y entre 0 e 150 segundos

# Adicionar rtulos aos eixos
plt.xlabel('Problema')
plt.ylabel('Tempo de Execuo (segundos)')

# Exibir o grfico
plt.tight_layout()  # Ajusta o layout para que tudo fique visvel
plt.legend()  # Exibe a legenda
plt.show()
```

```python
# Juntar os dados dos dois algoritmos com base no nome do problema
merged_df = pd.merge(chr_df[['nome_do_problema','caminho', 'fator_aproximacao']], tat_df[['nome_do_problema','caminho', 'fator_aproximacao']], on='nome_do_problema', suffixes=('_bb', '_tat'))

# Calcular a diferena no fator de aproximao
merged_df['diferenca_fator_aproximacao'] = abs(merged_df['fator_aproximacao_bb'] - merged_df['fator_aproximacao_tat'])

# Verificar as primeiras linhas do dataframe mesclado
merged_df.head()

# Remover a primeira linha do merged_df
merged_df = merged_df.iloc[1:].reset_index(drop=True)

# Verificar as primeiras linhas do dataframe atualizado
merged_df.head()
```

```python
# Problema com a maior diferena no fator de aproximao
max_diff_problem = merged_df.loc[merged_df['diferenca_fator_aproximacao'].idxmax()]

# Problema com a menor diferena no fator de aproximao
min_diff_problem = merged_df.loc[merged_df['diferenca_fator_aproximacao'].idxmin()]

# Exibir os problemas selecionados
print("Problema com a maior diferena no fator de aproximao:")
print(max_diff_problem)

print("\nProblema com a menor diferena no fator de aproximao:")
print(min_diff_problem)
```

```python
import ast

# Converter a string para lista de inteiros
caminho_bb = ast.literal_eval(max_diff_problem['caminho_bb'])
caminho_tat = ast.literal_eval(max_diff_problem['caminho_tat'])

# Garantir que os valores sejam inteiros
caminho_bb = [int(x) for x in caminho_bb]
caminho_tat = [int(x) for x in caminho_tat]

# Criar o grfico
plt.plot(caminho_bb, color='blue')
plt.plot(caminho_tat, color='red')

# Adicionar ttulo e rtulos
plt.title(f'Caminhos do Problema {max_diff_problem["nome_do_problema"]} com Maior Diferena no Fator de Aproximao')
plt.xlabel('Passos')
plt.ylabel('Nos acessados')

# Exibir o grfico
plt.show()
```
